## 5. Run Training

**Cú pháp:** `./autoencoder_cpu <data_path> [epochs] [batch_size] [max_samples] [optimizer]`

| Tham số | Mô tả | Mặc định |
|---------|-------|----------|
| data_path | Đường dẫn CIFAR-10 | ./cifar-10-batches-bin |
| epochs | Số epochs | 3 |
| batch_size | Kích thước batch | 32 |
| max_samples | Giới hạn samples (0=ALL) | 0 |
| optimizer | adam hoặc sgd | adam |

```python
# Test nhanh: 100 samples, 2 epochs, Adam
!./autoencoder_cpu ./cifar-10-batches-bin 2 32 100 adam
```

```python
# So sánh SGD
!./autoencoder_cpu ./cifar-10-batches-bin 2 32 100 sgd
```
